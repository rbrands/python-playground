```python
# Datenanalyse mit pandas und matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Beispieldaten
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Alter': [25, 30, 35, 28],
    'Punktzahl': [85, 92, 78, 88]
}

df = pd.DataFrame(data)
print(df)

# Einfaches Diagramm
plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Punktzahl'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Punktzahl')
plt.title('Punktzahl nach Person')
plt.show()
```
